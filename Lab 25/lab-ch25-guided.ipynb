### Q1: What worked well with AI-assisted development?

### Q2: What did you have to fix manually?

*Think about: layout bugs, incorrect assumptions, styling mismatches, missing features, hallucinated APIs.*

**Your answer:**

The biggest issues were visual/layout problems that required multiple iterations. The AeroNU rocket section was the most stubborn — Claude Code initially split it into an awkward two-panel layout (dark left, image right) instead of a full-bleed background, and it cropped the rocket image so the nose cone and launch flame were cut off. I had to go through 3-4 prompt iterations to get the image positioning, gradient overlay subtlety, and section height correct. ### Q3: How does AI-assisted development compare to writing code entirely yourself?

*Think about: the HADI loop from Chapter 25 (Hypothesis, Action, Data, Insight). Did you follow this pattern? Where did you need domain knowledge that the AI lacked?*

**Your answer:**

The process naturally followed the HADI loop. My hypothesis was that a comprehensive initial prompt with design references, resume content, and specific section descriptions would produce a near-complete site. The action was feeding that prompt to Claude Code. The data was the rendered output at localhost:3001, which revealed gaps — the rocket section layout was wrong, text needed recruiter-optimization, card spacing was off. The insight was that AI is excellent at generating structure and boilerplate but struggles with visual judgment calls like image cropping, gradient subtlety, and spatial balance. Domain knowledge I brought that the AI lacked included: understanding what PM recruiters actually scan for on a portfolio (which led to rewriting the bio to lead with user-obsession and end-to-end ownership rather than technical skills), knowing that "ML Engineer" in my subtitle would confuse my PM positioning, and recognizing that the AeroNU section needed emotional storytelling rather than just technical specs.

### Q4: Prompt engineering lessons

*What makes a good prompt for Claude Code? What makes a bad one? Give one specific example of a prompt you improved.*

**Your answer:**

Good prompts are specific about the *what* and clear about the *why*. My initial portfolio prompt worked well because it included concrete design references (samkolder.com, richeng.info), exact color values, specific section structures with real content, and technical requirements (Next.js App Router, Tailwind, Framer Motion). Bad prompts are vague about visual intent — saying "fix the layout" without specifying what's wrong. A specific example: my first attempt at fixing the rocket section was "make the image a full-bleed background," which produced the two-panel split. The improved prompt specified: "Make the rocket image a FULL-BLEED background behind the entire section — NOT split into two halves. The gradient overlay should be SUBTLE (40-50% opacity on the left, fading to transparent). The full rocket from nose to flame must be visible." The difference was being explicit about what I *didn't* want, providing opacity percentages, and describing the exact visual outcome rather than just the technique.

### Q5: If you were hiring a junior data scientist, would their portfolio site influence your evaluation? Why or why not?

*Connect to the "signaling" concept from labor economics — what does a portfolio signal that a resume alone does not?*

**Your answer:**

Absolutely — a portfolio is a costly signal in the Spence signaling model sense. A resume lists credentials (education, GPA, job titles), which are easy to verify but tell you little about how someone thinks or works. A portfolio signals execution ability, taste, and initiative — qualities that are hard to fake and expensive to produce in terms of time and effort. For a PM or data scientist specifically, a portfolio demonstrates that you can ship a complete product end-to-end, not just write code in isolation. It also reveals personality and interests, which matter for culture fit — my wildlife photography and conservation work, for example, signal curiosity and storytelling ability that a bullet point on a resume never could. The fact that I used AI tools to build it doesn't diminish the signal; if anything, it demonstrates a modern PM skill — knowing how to direct AI effectively to produce quality output, which is increasingly what the job actually looks like.
